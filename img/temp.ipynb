```python

import numpy as np
import numpy as np
from scipy.interpolate import make_interp_spline
import matplotlib.pyplot as plt
 
# Dataset
x = np.array([0, 2, 3, 4, 5, 6, 7, 8])
y = np.array([0, 87, 85, 1, 1, 1, 1, 1])
 
X_Y_Spline = make_interp_spline(x, y)
 
# Returns evenly spaced numbers
# over a specified interval.
X_ = np.linspace(x.min(), x.max(), 1000)
Y_ = X_Y_Spline(X_)
 
# Plotting the Graph
plt.plot(X_, Y_)
# plt.title("Plot Smooth Curve Using the scipy.interpolate.make_interp_spline() Class")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the x values
x = np.linspace(1, 10, 100)

# Define the y values using logarithmic function
y = np.log10(x)

# Create a new figure
fig, ax = plt.subplots()

# Plot the logarithmic curve
ax.plot(x, y)

# Set the x and y labels
ax.set_xlabel('Effort' , fontweight='bold')
ax.set_ylabel('Defects Detected (%)', fontweight='bold')

ax.set_xticklabels([])
ax.set_yticklabels([])

# Show the plot
plt.show()
plt.savefig("eff vs defects.png")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data
effort = np.arange(1, 10, 1)
defects_detected = np.power(2, effort) - 1

# Create a new figure
fig, ax = plt.subplots()

# Plot the logarithmic curve
ax.semilogx(effort, defects_detected, marker='o')

# Set the x and y labels in bold
ax.set_xlabel('Effort', fontweight='bold')
ax.set_ylabel('Defects Detected (100%)', fontweight='bold')
fig.savefig('eff_vs_defects.png')
# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data
effort = np.arange(1, 10, 1)
defects_detected = np.power(2, effort) - 1

# Create a new figure
fig, ax = plt.subplots()

# Plot the logarithmic curve
ax.semilogx(effort, defects_detected, marker='o')

# Set the x and y labels in bold
ax.set_xlabel('Effort', fontweight='bold')
ax.set_ylabel('Defects Detected (100%)', fontweight='bold')

# Remove x ticks and y ticks
ax.tick_params(axis='x', which='both', bottom=False, top=False)
ax.tick_params(axis='y', which='both', left=False, right=False)

# Save the plot as a PNG file
fig.savefig('logarithmic_curve.png')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a new figure
fig, ax = plt.subplots()

# Plot a half-smooth curve
ax.plot(x, y, label='Half-smooth Curve', linestyle='--', linewidth=2, alpha=0.7)

# Plot a straight plot
ax.plot(x, x, label='Straight Plot', linestyle='-', linewidth=2, alpha=0.7)

# Set the x and y labels
ax.set_xlabel('X')
ax.set_ylabel('Y')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data
x = np.linspace(0, 10, 100)
y1 = -x + 10
y2 = np.piecewise(x, [x < 5, x >= 5], [lambda x: 2*x, lambda x: 10])

# Create a new figure
fig, ax = plt.subplots()

# Plot a straight anti-plot
ax.plot(x, y1, label='Straight Anti-plot', linestyle='-', linewidth=2, alpha=0.7)

# Plot a straight plot that is flat after some time
ax.plot(x, y2, label='Straight plot with flat portion', linestyle='-', linewidth=2, alpha=0.7)

# Set the x and y labels
ax.set_xlabel('X')
ax.set_ylabel('Y')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data
x = np.linspace(0, 10, 100)
y1 = 2*x + 3
y2 = np.piecewise(x, [x < 5, x >= 5], [lambda x: 2*x+3, 13])

# Create a new figure
fig, ax = plt.subplots()

# Plot a straight line for the entire plot
ax.plot(x, y1, label='Straight Plot', linestyle='-', linewidth=2)

# Plot a line that is straight at the beginning and flat after some time
ax.plot(x, y2, label='Flat Plot', linestyle='--', linewidth=2)

# Set the x and y labels
ax.set_xlabel('X')
ax.set_ylabel('Y')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data
x = np.linspace(0, 10, 100)
y1 = 2*x + 1
y2 = np.zeros_like(x)
y2[:50] = 2*x[:50] + 1
y2[50:] = y2[49]

# Create a new figure
fig, ax = plt.subplots()

# Plot the first line (straight)
ax.plot(x, y1, label='Manual', linestyle='-', linewidth=2, alpha=0.7)

# Plot the second line (flat after some time with a smooth curve)
ax.plot(x, y2, label='Automated', linewidth=2, alpha=0.7)

# Set the x and y labels
ax.set_xlabel('Time')
ax.set_ylabel('Cumulative Costs ($)')

# Add a legend
ax.legend()
ax.set_xticklabels([])
ax.set_yticklabels([])
fig.savefig("Cost vs Time.png")
# Show the plot
plt.show()
```
