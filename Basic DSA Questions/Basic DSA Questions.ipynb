## Fibonacci Series



````python
# by using Recursion
def fibonacciSeries(n):
    if n == 0: return 0
    if n == 1: return 1
    return fibonacciSeries(n-1) + fibonacciSeries(n - 2)
````

````python
for i in range(8 + 1):
    print(fibonacciSeries(i),end=" ")
````

````python
# by using iterative approach
def fibonaccieSeries(n):
    a = 0
    b = 1
    print(a,end=" ")
    print(b, end=" ")
    if n == 0:
        print(a)
    elif n == 1:
        print(b)
    else:
        for i in range(2, n + 1):
            c = a + b
            a = b
            b = c
            print(b,end=" ")
````

````python
print(fibonaccieSeries(8))
````

````python
starPattern1(10)
````

````python
def starpattern2(n):
    for i in range(n):
        print("*"*i)
````

````python
starpattern2(10)
````

````python
def starpattern3(n):
    i = 1
    j = n - i # gaps
    i2 = 0
    while i<=n:
        print(" "*i2,"*"*1,"   "*j,"*"*1,end=" ")
        i += 1
        j -= 1
        i2 += 1
        print()
    
    i = n - i
    j = 0 # gaps
    i2 = n - i
    while j<=n:
        print(" "*i2,"*"*1,"   "*j,"*"*1,end="")
        i -= 1
        j += 1
        i2 -= 1
        print()
````

````python
# effiecient from above
def starpattern3(n):
    for i in range(n):
        for j in range(n):
            if i == j or i + j == n - 1:  # Diagonal conditions
                print("*", end="")
            else:
                print(" ", end="")
        print()

starpattern3(20)
````

````python
starpattern3(10)
````

## Palindrome

````python
def palindrome(s):
    i = 0
    j = len(s) - 1
    while i < j:
        if s[i] != s[j]:
            return False
        i += 1
        j -= 1
    return True
````

````python
palindrome("amma")
````

````python
palindrome("nitin")
````

````python
palindrome("lava")
````

## Program to count the occurence of each Character

* Using hash

````python
from collections import defaultdict
````

````python
def EachOccur(s):
    hashV = defaultdict(int)
    for i in s:
        hashV[i] += 1
    return hashV
````

````python
EachOccur("Hello")
````

````python
## without using collections library
def EachOccur(s):
    hashV = {}
    for i in s:
        if i not in hashV:
            hashV[i] = 1
        else:
            hashV[i] += 1
    return hashV
````

````python
EachOccur("hello")
````

## Valid Parenthesis

* Using Stack

````python
# support function
def isMatching(a, b):
    if (a == '(' and b == ')') or (a == '{' and b == '}') or (a == '[' and b == ']'):
        return True
    else:
        return False
def valParentesis(s):
    stack = []
    for i in s:
        if i in ['{','(','[']:
            stack.append(i)
        else:
            if not stack:
                return False
            if isMatching(stack[-1], i) == True:
                stack.pop()
            else:
                return False
    return True if not stack else False
````

````python
valParentesis('({[]})')
````

````python
valParentesis('{([)}')
````

## Find missing number in a set of `n` natural numbers

````python
# if i assume it is in a sorted order
def missingNumber(n):
    for i in range(len(n) - 1):
        if str(int(n[i]) + 1) != n[i+1]:
            return int(n[i]) + 1
````

````python
missingNumber("123467")
````

````python
# by using formula
def missingNumber(arr):
    n = len(arr) + 1
    total = n * (n + 1)//2
    sum_of_elements = sum(arr)
    return total - sum_of_elements
````

````python
missingNumber([1,2,4,5,6,7])
````

## FInd the 2nd largest number in an array 

`(w/o using sorting algo)`

```py
[1,3,2,5,4] --> 4

```

````python
def largest(arr, k):
    res = []
    while len(res) < k:
        maxArr = 0
        for i in range(len(arr)):
            if arr[i] > maxArr and arr[i] not in res:
                maxArr = arr[i]
        res.append(maxArr)
    return res[-1]
````

````python
largest([1,3,2,5,4], 2)
````

## Lowest and highest no in array

````python
def lowAndHigh(arr):
    low = float("inf")
    high = float("-inf")
    for i in arr:
        if i > high:
            high = i
        elif i < low:
            low = i
    return low, high
````

````python
lowAndHigh([4,2,5,87,1,0])
````

## Facorial Problem

### Using Recursion

````python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)
````

````python
factorial(5)
````

### Without using recursion

````python
def factorial1(n):
    res = 1
    for i in range(1,n+1):
        res *= i
    return res
````

````python
factorial1(5)
````

## Contains Duplicate

````python
def containsDuplicate(arr):
    hashV = defaultdict(int)
    for i in arr:
        hashV[i] += 1
        if hashV[i] > 1:
            return True
    return False
````

````python
containsDuplicate([1,2,3,4,5])
````

## Delete Duplicates from Array

````python
def DeleteDuplicates(arr):
    hashV = defaultdict(int)
    res = []
    for i in range(len(arr)):
        hashV[arr[i]] += 1
        if hashV[arr[i]] == 1:
            res.append(arr[i])
    return res
````

````python
DeleteDuplicates([1,2,3,3,4,5,6])
````

## Intersection of 2 arrays


```
arr1 = [1,2,3]
arr2 = [3,4,5]
```

````python
def intersection2Array(arr1, arr2):
    res = []
    for i in arr2:
        if i in arr1:
            res.append(i)
    return res

# TC: o(m * n)
# SC: o(n)
````

````python
intersection2Array([1,2,3],[3,4,5])
````

````python
## using set
def intesection2ArraySet(arr1, arr2):
    s_arr1 = set(arr1)
    res = [i for i in arr2 if i in s_arr1]
    return res

# TC: o(M + n)
# SC: o(n)
````

````python
intesection2ArraySet([1,2,3],[3,4,5])
````
