```python
import pandas as pd
```

```python
df = pd.read_csv("Book1.csv")
```

```python
df
```

```python
link = "https://leetcode.com/problems/"
```

```python
leetcode_links = []
```

```python
for i in df["name"]:
    k = ("-").join(i.split()[1:]).lower()
    print(k)
    leet_link = link + k
    print(leet_link)
    leetcode_links.append(leet_link)
```

```python
df["leetcode links"] = leetcode_links
```

```python
df
```

```python
df.to_csv("Book1_updated.csv", index=False)
```
